```python
#a ind vs aus ind as given 298 as traget as aus the rus for every 5 overs is given as follows
#india -[25,51,84,131,160,189,220,250,267,297]
#australia have complete 25 overs  -[15,41,94,110,151]
#plot over vs runs for both the countrys with india has blue line and austrila has yellow line....give a apporiate lables and legend and the marker for india is flag and marker for austrila is plus and take over from arange function and create arrays for runs


import numpy as np
import matplotlib.pyplot as plt
overs=np.arange(5,51,5)
overs1=np.arange(5,26,5)
i=np.array([25,51,84,131,160,189,220,250,267,297])
a=np.array([15,41,94,110,151])
plt.plot(overs,i,marker='>',label='india',color='b')
plt.plot(overs1,a,marker='+',label='aus',color='y')
plt.title("ind vs aus")
plt.xlabel("overs")
plt.ylabel("runs")
plt.legend()
plt.grid()

```
